```python
def busca(key,film_name):
    url = "https://api.themoviedb.org/3/search/movie?api_key="+key+"&language=en-US&query="+film_name
    res = requests.get(url)
    print(url)
    return res.json()
```

```python
df["popularity"]=""
```

```python
def actualiza_valor_fila(tabla,fila_valor,fila,columna,valor):
    tabla.loc[tabla[f"{fila_valor}"]== fila,columna]= valor
```

```python
for i in df["title"]:
    try:
        prueba1= busca('30d863c4e4e7628093e57a7c0240b87f',urllib.parse.quote(f"{i}"))
        popularity= prueba1["results"][0]["popularity"]
        #print(popularity)
        actualiza_valor_fila(df,"title",i,"popularity",popularity)
    except:
        actualiza_valor_fila(df,"title",i,"popularity",None)
```
